```python
# Imports libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#  Load the dataset
ds = pd.read_csv('House Price Prediction Dataset.csv')
ds
```

```python
#  Check for duplicates
ds.duplicated().sum()
ds.drop_duplicates(inplace=True)
```

```python
# 5. Check for missing values 
print( ds.isnull().sum())
```

```python
# Handle outliers
ds = ds[ds['Area'].between(400, 10000)]
ds = ds[ds['Price'].between(20000, 2000000)]
ds = ds[ds['Bedrooms'].between(1, 10)]
ds = ds[ds['Bathrooms'].between(1, 10)]
ds = ds[ds['Floors'].between(1, 5)]
ds = ds[ds['YearBuilt'] >= 1930]
```

```python
# check
ds.info()
ds.describe()
```

```python
# to save file
ds.to_csv('Cleaned_House_Price_Data.csv', index=False)
```
